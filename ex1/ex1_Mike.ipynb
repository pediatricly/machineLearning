This works by combining 2 functions:
[asarray](http://docs.scipy.org/doc/numpy/reference/generated/numpy.asarray.html) just takes array-compatible Python data objects (lists, tupules) and turns them into arrays.

[reshape](http://docs.scipy.org/doc/numpy/reference/generated/numpy.reshape.html) "ravels" an array into a scalar, preserving order from top to bottom. It then follows that order to fill out the dimensions you specify. (Wouldn't want to do this with multiple x's & y as this would ruin the information rows, but it works to convert scalar -> vector array.)
Of note, the arguments for reshape are the dimensions you want. -1 is a wildcard (I think). Can use one per argument set and so it just figures out how to lay out the data. In this example, -1 or 97 (# of rows) in that slot work equivalently, but anything else (0, 8, 96, 100...) produces an error because it can't fit the data given to the dimensions specified.

The [insert function](http://docs.scipy.org/doc/numpy/reference/generated/numpy.insert.html) works like this:

np.insert(startingArray, indexToStartInserting, valuesToInsert, axisAlongWhichToInsert)
 - indexToStartInserting can be an integer like this or can be a multidimensional index or a sequence for multiple inserts.
 - values just needs to be an array like object
 - axis is the axis to follow where zero is the first (axis along the first row) and 1 is the column axis (can do higher dimensions too). If left blank, the array is flattened to insert and you get a scalar back.

## Gradient Descent

```python
iter = 1500
alpha = 0.01
```

```python
def h(theta,X): #Linear hypothesis function
    return np.dot(X,theta) 
# np.dot returns the dot product of a matrix
# This is just a vectorized way to calculate hypotheses from vectors of x's and thetas. 
# It's a quicker way to represent multiplying the x's by the theta's and summing them into 
# h(x)

def computeCost(mytheta,X,y): #Cost function
    """
    theta_start is an n- dimensional vector of initial theta guess
    X is matrix with n- columns and m- rows
    y is a matrix with m- rows and 1 column
    """
    #note to self: *.shape is (rows, columns)
    return float((1./(2*m)) * np.dot((h(mytheta,X)-y).T,(h(mytheta,X)-y)))

#Test that running computeCost with 0's as theta returns 32.07:

initial_theta = np.zeros((X2.shape[1],1)) #(theta is a vector with n rows and 1 columns (if X has n features) )
print computeCost(initial_theta,X2,Y)
```

```python
#Actual gradient descent minimizing routine
def descendGradient(X, theta_start = np.zeros(2)):
    """
    theta_start is an n- dimensional vector of initial theta guess
    X is matrix with n- columns and m- rows
    """
    theta = theta_start
    jvec = [] #Used to plot cost as function of iteration
    thetahistory = [] #Used to visualize the minimization path later on
    for meaninglessvariable in xrange(iter):
        tmptheta = theta
        jvec.append(computeCost(theta,X,Y))
        # Buggy line
        #thetahistory.append(list(tmptheta))
        # Fixed line
        thetahistory.append(list(theta[:,0]))
        #Simultaneously updating theta values
        for j in xrange(len(tmptheta)):
            tmptheta[j] = theta[j] - (alpha/m)*np.sum((h(initial_theta,X) - Y)*np.array(X[:,j]).reshape(m,1))
        theta = tmptheta
    return theta, thetahistory, jvec
```

```python
#Actually run gradient descent to get the best-fit theta values
initial_theta = np.zeros((X2.shape[1],1))
theta, thetahistory, jvec = descendGradient(X2,initial_theta)

#Plot the convergence of the cost function
def plotConvergence(jvec):
    plt.figure(figsize=(10,6))
    plt.plot(range(len(jvec)),jvec,'bo')
    plt.grid(True)
    plt.title("Convergence of Cost Function")
    plt.xlabel("Iteration number")
    plt.ylabel("Cost function")
    dummy = plt.xlim([-0.05*iter,1.05*iter])
    #dummy = plt.ylim([4,8])


plotConvergence(jvec)
dummy = plt.ylim([4,7])
```

```python
#Plot the line on top of the data to ensure it looks correct
def myfit(xval):
    return theta[0] + theta[1]*xval
plt.figure(figsize=(10,6))
plt.plot(X2[:,1],Y[:,0],'ro',markersize=10,label='Training Data')
plt.plot(X2[:,1],myfit(X2[:,1]),'b-',label = 'Hypothesis: h(x) = %0.2f + %0.2fx'%(theta[0],theta[1]))
plt.grid(True) #Always plot.grid true!
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.legend()
```
