## Gradient Descent

```python
iter = 1500
alpha = 0.01
```

Below the basic h(theta) (the linear estimate based on whatever the guess of theta aka the betas) is computed. That's the def h(theta,X function).

And the cost over the entire training set for that estimate is computed. Apparently in linear algebra, you can compute that cost with dot( transpose(h(theta,X)-y) & h(theta,X)-y )
(Of course, those h(theta,X) are also dot products, so that whole thing is the dot product of 2 dot products.)
But, that's all just the Gaussian squared distance of each actual (x,y) in the training set from the estimated/predicted (x, h(x)) line. 
How the dot products and transpose do that I only sort of understand, but the concept is just an easier way to code ordinary least squares. (Well, the first step in OLS - this is just the cost for our first prediction. Calculating the cost here is to track the model.


All this is just the long way of coding your own simple gradient descent algorithm. There is, of course, a library that **implements this for you in Python - [scikit gradient descent](http://scikit-learn.org/stable/modules/sgd.html)**. Looks pretty simple and, indeed, with this course, I understand most of it.

```python
def h(theta,X): #Linear hypothesis function
    return np.dot(X,theta) 
# np.dot returns the dot product of a matrix
# This is just a vectorized way to calculate hypotheses from vectors of x's and thetas. 
# It's a quicker way to represent multiplying the x's by the theta's and summing them into 
# h(x)

def computeCost(mytheta,X,y): #Cost function
    """
    theta_start is an n- dimensional vector of initial theta guess
    X is matrix with n- columns and m- rows
    y is a matrix with m- rows and 1 column
    """
    #note to self: *.shape is (rows, columns)
    return float((1./(2*m)) * np.dot((h(mytheta,X)-y).T,(h(mytheta,X)-y)))

#Test that running computeCost with 0's as theta returns 32.07:

print X2.shape

initial_theta = np.zeros((X2.shape[1],1)) #(theta is a vector with n rows and 1 columns (if X has n features) )
print initial_theta
# print h(initial_theta,X2) #Show that h(x) returns a vector of h(x) predictions for every 
# set of Xs (in this case they're (1, xi)). These are all zero because initial_theta is (0,0)
# so h(theta,X) effectively does (0*1 + 0*xi)
print computeCost(initial_theta,X2,Y)
```

### 2.2.1 Update Equations
The objective of linear regression is to minimize the cost function:

`J(theta) = 1/2m * {sum from i = 1 to m of [h(x)-y]^2}`

where the hypothesis h(x) is given by the linear model:

`h(x) = theta-transpose * x = theta0 + theta1 * x`
    
Recall that the parameters of your model are the j values. These are the values you will adjust to minimize cost J(). One way to do this is to use the batch gradient descent algorithm. In batch gradient descent, each iteration performs the update

`j := j   (h(x(i))  y(i))x(i) (simultaneously update j for all j).`

The tmptheta[j] = theta[j] - (alpha/m)... line does this simultaneous update for all thetas:
- It's subtraction because you want to trace down the slope at that point
- Multiply by alpha to jump along at the learning rate
- Divide by m (size of training set) to average the errors over the training set
- np.sum((h(initial_theta,X) - Y) is the sum of the errors (not squared)
- np.array(X[:,j]).reshape(m,1)) multiplies by X which is the term that appears when you take the partial derivative of the squared errors with respect to X
- Multiply by 2 & divide by 2 - doesn't appear. Multiply by 2 from taking the derivative of the square. But we've already divided by 2 as in the first equation above because we're actually minimizing half the cost instead of the whole cost. But dividing by the scalar 2 doesn't change the minimization (still looking for 0) and doing so lets these 2s cancel out.

```python
#Actual gradient descent minimizing routine
def descendGradient(X, theta_start = np.zeros(2)):
    """
    theta_start is an n- dimensional vector of initial theta guess
    X is matrix with n- columns and m- rows
    """
    theta = theta_start
    jvec = [] #Used to plot cost as function of iteration
    thetahistory = [] #Used to visualize the minimization path later on
    for meaninglessvariable in xrange(iter):
        tmptheta = theta #Update theta from the last iteration
        jvec.append(computeCost(theta,X,Y)) #Store the cost for visualization only
        thetahistory.append(list(theta[:,0])) #Store the theta for visualization only
        #Simultaneously updating theta values
        #This is what actually runs the gradient descent on the cost function...
        #j the iteration var is index for the dimensions of x & theta 
        for j in xrange(len(tmptheta)):
            #Then here we update theta by the gradient * learning rate. Explanation above.
            tmptheta[j] = theta[j] - (alpha/m)*np.sum((h(initial_theta,X) - Y)*np.array(X[:,j]).reshape(m,1))
        # Watch theta converge from theta_start to eventual prediction:
        # print tmptheta
        theta = tmptheta
    return theta, thetahistory, jvec
# print m # m appears in the function, invoked as a global variable, same as alpha
```

```python
#Actually run gradient descent to get the best-fit theta values
initial_theta = np.zeros((X2.shape[1],1))
theta, thetahistory, jvec = descendGradient(X2,initial_theta)
print 'final theta: ', theta

#Plot the convergence of the cost function
def plotConvergence(jvec):
    plt.figure(figsize=(10,6))
    plt.plot(range(len(jvec)),jvec,'bo')
    plt.grid(True)
    plt.title("Convergence of Cost Function")
    plt.xlabel("Iteration number")
    plt.ylabel("Cost function")
    dummy = plt.xlim([-0.05*iter,1.05*iter])
    #dummy = plt.ylim([4,8])


plotConvergence(jvec)
dummy = plt.ylim([4,7])
```

```python
#Plot the line on top of the data to ensure it looks correct
def myfit(xval):
    return theta[0] + theta[1]*xval
plt.figure(figsize=(10,6))
plt.plot(X2[:,1],Y[:,0],'ro',markersize=10,label='Training Data')
plt.plot(X2[:,1],myfit(X2[:,1]),'b-',label = 'Hypothesis: h(x) = %0.2f + %0.2fx'%(theta[0],theta[1]))
plt.grid(True) #Always plot.grid true!
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.legend()
```
